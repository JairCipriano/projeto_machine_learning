```python
import random
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
```

```python
projetos = [f'Projeto {i+1}' for i in range(20)]
localizacoes = [f'Local {random.randint(1, 100)}' for _ in range(20)]
custos = [random.randint(100000, 1000000) for _ in range(20)]
datas_inicio = [datetime.now() - timedelta(days=random.randint(30, 365)) for _ in range(20)]
datas_termino = [data + timedelta(days=random.randint(30, 365)) for data in datas_inicio]
status = random.choices(['Em andamento', 'Concluído', 'Cancelado'], k=20)
responsaveis = [f'Engenheiro {random.randint(1, 10)}' for _ in range(20)]
tipos_construcao = random.choices(['Residencial', 'Comercial', 'Industrial'], k=20)
areas = [random.randint(50, 500) for _ in range(20)]
andares = [random.randint(1, 10) for _ in range(20)]

# Criar o DataFrame
df_construcao = pd.DataFrame({
    'Nome do Projeto': projetos,
    'Localização': localizacoes,
    'Custo (R$)': custos,
    'Data de Início': datas_inicio,
    'Data de Término': datas_termino,
    'Status': status,
    'Responsável': responsaveis,
    'Tipo de Construção': tipos_construcao,
    'Área (m²)': areas,
    'Número de Andares': andares
})
df_construcao
```

```python
df_construcao.to_csv('teste1',index=False)
```

```python
df = pd.DataFrame({ 'A': [1, 2, 3, 4], 
                    'B': [10, 20, 30, 40],
                    'C': [100, 200, 300, 400]}, 
                     index=['Linha 1', 'Linha 2', 'Linha 3', 'Linha 4'])
```

```python
df
```

```python
def nossa_soma(series):
    return series.sum()
```

```python
df['Soma do A + B + C'] = df.apply(nossa_soma, axis=1)
df
```

```python
df.loc['linha 5'] = df.apply(nossa_soma, axis=0)
df
```

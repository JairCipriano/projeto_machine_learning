```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

```

```python
import yfinance as yf
import datetime

```

```python
%matplotlib inline
```

```python
biticoin = yf.download('BTC-USD',
                              start = datetime.datetime(2018,1,1),
                              end = datetime.datetime(2022,4,1))
```

```python
biticoin
```

```python
biticoin['Close'].plot(figsize=(16,10))
```

```python
btc = biticoin.loc['2021-01': '2022-04'].copy()
btc
```

```python
btc['MMS30'] = btc['Close'].rolling(window= 30).mean()
```

```python
btc.head(31)
```

```python
fig, ax = plt.subplots(figsize = (14,8))
ax.plot(btc['Close'],label = 'biticoin')
ax.plot(btc['MMS30'],label = 'MMS 30')
ax.plot(btc['Close'].rolling(window =  15).mean(),label = 'MMS 15')
#.rolling - referente ao fechamento
#window Ã© o periodo de tempo que vocÃª quer
#.mean() Ã© a media
#label Ã© o nome que vocÃª quer chamar extamente aquelka linha do grafico

plt.legend()
#ple.legend Ã© usado para adicionar a caixinha da legenda
plt.show()
#plt.show Ã© uxado para impremir o grafico na tela
```

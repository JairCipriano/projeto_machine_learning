```python
carros = pd.DataFrame(
    np.random.randint(3000, 25000, size=(8, 12)),  # Ajustando os limites para que o primeiro seja menor que o segundo
    index=['fiat', 'peugeot', 'volkswagen', 'ford', 'nissan', 'toyota', 'honda', 'chevrolet'],  # Corrigindo o nome 'pegeot' para 'peugeot' e 'volkswagem' para 'volkswagen'
    columns=['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']  # Corrigindo 'des' para 'dez'
)
carros
```

```python
carros['total'] = carros.sum(axis=1)
carros
```

```python
#ao usar a função  sort_values a organizaão fica em ordem crescente
#Vamos organizar pela coluna total
carros.sort_values (by='total')
```

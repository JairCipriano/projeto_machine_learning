```python


plt.figure(figsize=(12, 5))  # Define o tamanho da figura

# Primeiro gráfico
plt.subplot(1, 2, 1)  # 1 linha, 2 colunas, posição 1
plt.plot(carros4['volkswagen'], color='blue', label='Volkswagen')
plt.xticks(rotation=90)
plt.title('Volkswagen')
plt.legend()

# Segundo gráfico
plt.subplot(1, 2, 2)  # 1 linha, 2 colunas, posição 2
plt.plot(carros4['ford'], color='red', label='Ford')
plt.xticks(rotation=90)
plt.title('Ford')
plt.legend()

plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```
